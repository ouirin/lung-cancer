```python
import pandas
import settings
import helpers
import random
import os
import glob
import ntpath
import math
```

```python
# Output from Scanner
O_dir = settings.BASE_DIR_SSD + "NMS_Result/CNN_10/"

count=0

threshold=0.9

for patient_index, csv_file in enumerate(glob.glob(O_dir + "*.csv")):

    patient_id = ntpath.basename(csv_file).replace(".csv", "") 
    
    df_O = pandas.read_csv(csv_file)
    
    pos_lines = []
    
    for index_O, row_O in df_O.iterrows():
        
        temp = row_O["nodule_chance"]
        
        if temp>threshold:
            pos_lines.append(row_O)
            count+=1
    
    print(count)
    #print(pos_lines)
    df_annos = pandas.DataFrame(pos_lines, columns=["anno_index", "coord_x", "coord_y", "coord_z", "nodule_chance"])
    
    dst_dir="NMS_Result/CNN_10_" + str(int(threshold * 10)) + "/"
    
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
    
    df_annos.to_csv(dst_dir + patient_id + "_annos_pos_lidc.csv", index=False)
    
print(count)
```

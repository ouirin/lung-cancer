```python
def predict_cubes():
    
    csv_target_path = "D:/jupyter-notebook/LungCancerPredict/NDSB_Annotations_From_CNN/ResNet/out.csv"
    #计时器
    sw = helpers.Stopwatch.start_new()
    
    #读取model
    model_path_texture="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_texture.hd5"
    model_path_subtlety="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_subtlety.hd5"
    model_path_spiculation="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_spiculation.hd5"
    model_path_sphericiy="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_sphericiy.hd5"
    model_path_margin="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_margin.hd5"
    model_path_malignancy="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_malignancy.hd5"
    model_path_lobulation="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_lobulation.hd5"
    model_path_diamter="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_diamter.hd5"
    model_path_calcification="D:/jupyter-notebook/LungCancerPredict/models/attribute_models_CNN_v1/out_calcification.hd5"


    model_texture       = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_texture)
    model_subtlety      = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_subtlety)
    model_spiculation   = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_spiculation)
    model_sphericiy     = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_sphericiy)
    model_margin        = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_margin)
    model_malignancy    = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_malignancy)
    model_lobulation    = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_lobulation)
    model_diamter       = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_diamter)
    model_calcification = Attribute_Predictor_CNN.get_net(load_weight_path=model_path_calcification)
    
    ##############################################################################################
    #读取所有img_paths
    samples = glob.glob(settings.BASE_DIR_SSD + "NDSB_Annotations_From_CNN/ResNet/cube/*.png")
    print("Samples Quantity: ", len(samples))
    
    random.shuffle(samples)
    
    img_paths=[]
    
    for index, sample_path in enumerate(samples):    

        file_name = ntpath.basename(sample_path)

        img_paths.append(sample_path)
            
    ##############################################################################################
    #读取所有imgs
    out_csv = []
    img_list = []
    info_list = []
    CROP_SIZE = CUBE_SIZE
    batch_size = 32
    done_count = 0  
    
    for img_idx, img_item in enumerate(img_paths):
        
        print(img_idx)
        
        ##############################################################################################
        #获取nodule的其他信息
        base_path = ntpath.basename(img_item)
        parts = base_path.split("_")
        chance = parts[-3]
        z = parts[-4]
        y = parts[-5]
        x = parts[-6]
        patient_id = parts[-8]
        info_list.append((patient_id,x,y,z,chance))

        ###############################################################################################
        #获取nodule的图片信息
        cube_image = helpers.load_cube_img(img_item, 8, 8, 64)

        current_cube_size = cube_image.shape[0]

        indent_x = (current_cube_size - CROP_SIZE) / 2
        indent_y = (current_cube_size - CROP_SIZE) / 2
        indent_z = (current_cube_size - CROP_SIZE) / 2

        indent_x = int(indent_x)
        indent_y = int(indent_y)
        indent_z = int(indent_z)

        #截取到crop_size大小的cube
        cube_image = cube_image[indent_z:indent_z + CROP_SIZE, indent_y:indent_y + CROP_SIZE, indent_x:indent_x + CROP_SIZE]

        if CROP_SIZE != CUBE_SIZE:
            cube_image = helpers.rescale_patient_images2(cube_image, (CUBE_SIZE, CUBE_SIZE, CUBE_SIZE))

        assert cube_image.shape == (CUBE_SIZE, CUBE_SIZE, CUBE_SIZE)

        #3D卷积的正规化 32*32*32
        img3d = prepare_image_for_net3D(cube_image)

        #添加数据
        img_list.append(img3d)
        
        ##############################################################################################
        #凑够32个，就开始对cube的属性进行预测
        done_count += 1

        if len(img_list)% batch_size==0 or done_count==len(img_paths):

            batch_data = numpy.vstack(img_list)
            p_diamter = model_diamter.predict(batch_data, batch_size=batch_size)
            p_malignancy = model_malignancy.predict(batch_data, batch_size=batch_size)
            p_spiculation = model_spiculation.predict(batch_data, batch_size=batch_size)
            p_lobulation = model_lobulation.predict(batch_data, batch_size=batch_size)
            p_sphericiy = model_sphericiy.predict(batch_data, batch_size=batch_size)
            p_margin = model_margin.predict(batch_data, batch_size=batch_size)
            p_subtlety = model_subtlety.predict(batch_data, batch_size=batch_size)
            p_texture = model_texture.predict(batch_data, batch_size=batch_size)
            p_calcification = model_calcification.predict(batch_data, batch_size=batch_size)
            
            for i in range(len(p_diamter)):
                
                #pre info
                diamter = str(p_diamter[i][0])
                malignancy = str(p_malignancy[i][0])
                spiculation = str(p_spiculation[i][0])
                lobulation = str(p_lobulation[i][0])
                sphericiy = str(p_sphericiy[i][0])
                margin = str(p_margin[i][0])
                subtley = str(p_subtlety[i][0])
                textrue = str(p_texture[i][0])
                calcification = str(p_calcification[i][0])
                
                #other info
                patient_id=info_list[i][0]
                x = info_list[i][1]
                y = info_list[i][2]
                z = info_list[i][3]
                chance = info_list[i][4]
                
                csv_line = [patient_id, x, y, z, chance,diamter,malignancy,spiculation,lobulation,sphericiy,margin,subtley,textrue,calcification]
                print(csv_line)
                out_csv.append(csv_line)
            
            img_list = []
            info_list = []
    
    ###########################################################################################
    #写成csv文件
    df = pandas.DataFrame(out_csv, columns=["patient_id", "coord_x", "coord_y", "coord_z", "nodule_chance","diamter","malignancy","spiculation","lobulation","sphericiy","margin","subtley","textrue","calcification"])
    df.to_csv(csv_target_path, index=False)

    ##############################################################################################
    print("Done in : ", sw.get_elapsed_seconds(), " seconds")
    
```

```python
def prepare_image_for_net3D(img):
    img = img.astype(numpy.float32)
    img -= MEAN_PIXEL_VALUE
    img /= 255.
    img = img.reshape(1, img.shape[0], img.shape[1], img.shape[2], 1)
    return img
```

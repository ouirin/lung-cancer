```python
def calc_texture_class(texture):
    
    threshold = [7/3,11/3]
   
    if texture>=threshold[0] and texture<=threshold[1]:
        vclass = 1
    elif texture>threshold[1]:
        vclass = 2
    else:
        vclass = 0
 
    return vclass
```
